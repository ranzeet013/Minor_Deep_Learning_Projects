```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
x = np.linspace(0, 50, 501)
y = np.cos(x)
```

```python
dataframe = pd.DataFrame(data = y, 
                         index = x, 
                         columns = ['Cosine'])
```

```python
dataframe.head()
```

```python
dataframe.tail()
```

```python
test_precent = 0.1
```

```python
len(dataframe)*test_precent
```

```python
test_point = np.round(len(dataframe)*test_precent)
```

```python
test_point
```

```python
test_index = int(len(dataframe) - test_point)
```

```python
test_index
```

```python
train_data = dataframe.iloc[:test_index]
test_data = dataframe.iloc[test_index:]
```
